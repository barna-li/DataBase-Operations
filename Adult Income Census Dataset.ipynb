## INNER JOIN THE TWO TABLES CENSUS & CENSUS2

```python
data = pd.read_sql_query("select IMP.census.education , IMP.census.workclass , IMP.census.age from IMP.census inner join IMP.census2 on IMP.census.education =IMP.census2.education",mydb)
data
```

## LEFT JOIN THE TWO TABLES CENSUS & CENSUS2

```python
data = pd.read_sql_query("select IMP.census.education , IMP.census.workclass , IMP.census.age, IMP.census.income from IMP.census left join IMP.census2 on IMP.census.education =IMP.census2.education",mydb)
data
```

## RIGHT JOIN THE TWO TABLES CENSUS & CENSUS2

```python
data = pd.read_sql("select IMP.census.education , IMP.census.workclass , IMP.census.age, IMP.census.income from IMP.census right join IMP.census2 on IMP.census.education =IMP.census2.education",mydb)
data
```
